```python
#import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Salary_Data.csv')  #import data file
```

```python
x = df.iloc[:,:-1].values #seprating DV
```

```python
y = df.iloc[:,1].values #seprating IV
```

```python
#split data into training and test data
#from sklearn.model_selection   import train_test_split
#train_test_split(x,y,test_size = 0.2, random_state = 0)
#test_size = 0.2 means it divide data into 80:20 train:test
#random_state prevent data shuffel, we will observe same data in traning and test
# split data into 4 variables
#x_train,y_train,x_test,y_test
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2, random_state =0)
```

```python
y_train.shape
```

```python
#from sklearn.linear_model import LinearRegression
# variable = LinearRegression()
#fit() take traing data and fit in model and call the actual linear regression algo

from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(x_train,y_train)
```

```python
#call predic method -> variable.predict(x_test)

y_predict =  reg.predict(x_test)
```
